scatter(x,y,s,c) s=size; c=color

```python
import matplotlib.pyplot as plt

plt.scatter(3,3)
plt.savefig('scatter01.jpg',bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt

xx = [1, 2, 3, 4, 5, 6]
yy = [3, 5, 2, 8, 7, 4]
plt.scatter(xx,yy,24,'r')
plt.xlim([min(xx)-1, max(xx)+1]) # x-axis([min, max])
plt.ylim([min(yy)-1, max(yy)+1]) # y-axis([min, max])
plt.savefig('scatter02.jpg',bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt

xx = [1, 2, 3, 4, 5, 6]
yy = [3, 5, 2, 8, 7, 4]
plt.scatter(xx,yy,24, 'r')
plt.xlim([min(xx)-1, max(xx)+1])
plt.ylim([min(yy)-1, max(yy)+1])
plt.grid()  # grid
plt.savefig('scatter03.jpg',bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt

xx = [1, 2, 3, 4, 5, 6]
yy = [3, 5, 2, 8, 7, 4]
plt.scatter(xx,yy,30, 'b')
plt.xlim([min(xx)-1, max(xx)+1])
plt.ylim([min(yy)-1, max(yy)+1])
plt.grid(color='r', linestyle=':', linewidth=2)
plt.savefig('scatter04.jpg',bbox_inches='tight')
plt.show()
```

Bar(x,y,width)

```python
import matplotlib.pyplot as plt

num = [10, 33, 25]
x = range(len(num))
plt.bar(x,num, 0.3)
plt.savefig('bar01.jpg',bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt

num = [10, 33, 25]
x = range(len(num))
plt.bar(x,num, 0.3)
plt.ylabel('The number of people')
plt.xticks(x, ('Peter', 'Bob', 'Ricky')) # X ticks
plt.yticks(range(0, 35, 4))
plt.savefig('bar02.jpg',bbox_inches='tight')
plt.show()
```

hist(x, bins, color, options ...)

```python
import matplotlib.pyplot as plt
from random import randint

times = 200
max_int = 10
rand_num = []
for i1 in range(times):
    rand_num.append(randint(1,max_int))
    
h = plt.hist(rand_num,4)
print('Y axis: ', h[0])
print('X axis: ', h[1])
plt.savefig('hist01.jpg',bbox_inches='tight')
plt.show()
```

pie(x, options...)

```python
import matplotlib.pyplot as plt

routine = ['Sleep', 'Play', 'Work', 'Eat', 'Others']
time = [7.5, 2.5, 9.5, 2.0, 2.5]
plt.pie(time, labels = routine)
plt.savefig('pie01.jpg',bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt

routine = ['Sleep', 'Play', 'Work', 'Eat', 'Others']
time = [7.5, 2.5, 9.5, 2.0, 2.5]
plt.pie(time, labels = routine, explode = (0,0.2,0,0,0),\
    autopct="%5.2f%%")
plt.savefig('pie02.jpg',bbox_inches='tight')
plt.show()
```

subplot(rows, cols, fig #)

```python
import matplotlib.pyplot as plt

xx = [1, 2, 3, 4 , 5, 6]
y1 = [1, 4, 9, 16, 25, 36]
y2 = [2, 7, 12, 9, 24, 15]

plt.subplot(2,1,1)
plt.plot(xx, y1, 'r-*')
plt.subplot(2,1,2)
plt.plot(xx,y2,'b--o')
plt.savefig('subplot01.jpg',bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt

xx = [1, 2, 3, 4 , 5, 6]
y1 = [1, 4, 9, 16, 25, 36]
y2 = [2, 7, 12, 9, 24, 15]

plt.subplot(1,2,1)
plt.plot(xx, y1, 'r-*')
plt.subplot(1,2,2)
plt.plot(xx,y2,'b--o')
plt.savefig('subplot02.jpg',bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt

xx = [1, 2, 3, 4 , 5, 6]
y1 = [1, 4, 9, 16, 25, 36]
y2 = [2, 7, 12, 9, 24, 15]

plt.subplot(2,3,1)
plt.plot(xx, y1, 'r-*')
plt.subplot(2,3,2)
plt.plot(xx,y1,'b-*')
plt.subplot(2,3,3)
plt.plot(xx, y1, 'k-*')
plt.subplot(2,3,4)
plt.plot(xx,y2,'r--o')
plt.subplot(2,3,5)
plt.plot(xx, y2, 'b--o')
plt.subplot(2,3,6)
plt.plot(xx,y2,'k--o')
plt.savefig('subplot03.jpg',bbox_inches='tight')
plt.show()
```
