matplotlib examples

```python
import matplotlib.pyplot as plt  # import pyplot as plt
num_sqs = [1, 4, 9, 16, 25, 36]
plt.plot(num_sqs) # plot values
plt.axis([0, 5, 0, 36]) # x: 0-5 & y: 0-36
plt.show()  # 
```

```python
num = [1, 2, 3, 4, 5, 6]  # numbers
plt.plot(num, num_sqs, color='red', linewidth=4)
plt.axis([1, 6, 0, 36])
plt.show()
```

```python
plt.plot(num, num_sqs, color='skyblue', linewidth=4)
plt.axis([1, 6, 0, 36])
plt.show()
```

```python
plt.plot(num, num_sqs, color =(101/255,107/255,255/255,0.5), linewidth=4)
plt.axis([1, 6, 0, 36])
plt.show()
```

```python
plt.plot(num, num_sqs,'r-',lw=2)
plt.axis([1, 6, 0, 36])
plt.title('Squares of Numbers', fontsize = 24)
plt.xlabel('Numbers')
plt.ylabel('Squares')
plt.show()
```

```python
num_test = [3, 7, 25, 12, 33, 6]
plt.plot(num, num_sqs, num, num_test)
plt.axis([1, 6, 0, 36])
plt.title('Test', fontsize = 24)
plt.xlabel('Numbers')
plt.ylabel('Values')
plt.show()
```

```python
plt.plot(num, num_sqs, 'r-', num, num_test,'b:')
plt.axis([1, 6, 0, 36])
plt.title('Test', fontsize = 24)
plt.xlabel('Numbers')
plt.ylabel('Values')
plt.show()
```

```python
plt.plot(num, num_sqs, color = 'gold', marker = '.', markersize = 14)
plt.plot(num, num_test,'k', marker = '*', markersize = 14)
plt.axis([1, 6, 0, 36])
plt.title('Test', fontsize = 24)
plt.xlabel('Numbers')
plt.ylabel('Values')
plt.show()
```

```python
plt.plot(num, num_sqs, 'b-o',label ='squares')
plt.plot(num, num_test,'k-s',label = 'test')
plt.legend()
plt.axis([1, 6, 0, 36])
plt.title('Test', fontsize = 24)
plt.xlabel('Numbers')
plt.ylabel('Values')
plt.show()
```

```python
plt.plot(num, num_sqs, 'b-o',label ='squares')
plt.plot(num, num_test,'k-s',label = 'test')
plt.legend(loc=4)
plt.axis([1, 6, 0, 36])
plt.title('Test', fontsize = 24)
plt.xlabel('Numbers')
plt.ylabel('Values')
plt.show()
```

```python
plt.plot(num, num_sqs, 'b-o',label ='squares')
plt.plot(num, num_test,'k-s',label = 'test')
plt.legend(loc=6)
plt.axis([1, 6, 0, 36])
plt.title('Test', fontsize = 24)
plt.xlabel('Numbers')
plt.ylabel('Values')
plt.savefig('/Users/rick/Desktop/fig_test.jpg',bbox_inches='tight')
plt.show()
```
